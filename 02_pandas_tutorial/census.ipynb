```python
%matplotlib inline

import pandas as pd
import numpy as np
```

```python
col_names = []
with open('census/census-income.names') as txt_file:
    lines = txt_file.readlines()
    for i,line in enumerate(lines):
        if 23 <= i <= 68:
            name = line.split('\t\t')[0].split('| ')[1]
            col_names.append(name)
col_names
```

```python
census_data = pd.read_csv("census/census-income.data", header=None, names=col_names)
census_data.head(4)
```

```python
census_test = pd.read_csv("census/census-income.test", header=None, names=col_names)
census_test.head(4)
```
