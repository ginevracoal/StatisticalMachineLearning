## Applying functions to Series and DataFrames

```python
# change column name
thanks["gender"] = thanks["What is your gender?"]
```

```python
def gender_code(gstr):
    if (pd.isnull(gstr)):
        return gstr
    elif gstr == "Female":
        return 1
    else:
        return 0
    
thanks["gender"].apply(gender_code)
thanks["gender"].value_counts(dropna=False)
```
